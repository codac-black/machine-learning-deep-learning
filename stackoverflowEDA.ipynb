```python
# check for missing values
def check_dataset(data):
  print(f'''
  missing values: {data.isna().sum().sum()}
  % missing values: {np.round(data.isna().sum().sum() / data.size * 100, 2)}%
  Size of data: {data.size}
  Shape of data: {data.shape}
  ''')

check_dataset(df)
```

```python
import plotly.graph_objects as go


# Calculate the top 25 countries
top_countries = df["Country"].value_counts().head(25)

rainbow_colors = ['#FF0000', '#FF7F00', '#FFFF00', '#00FF00', '#0000FF', '#4B0082', '#9400D3', '#FF4500', '#FFD700', '#008000',
                  '#0000FF', '#000080', '#4B0082', '#800080', '#FF00FF', '#800000', '#008080', '#00FFFF', '#808000', '#808080',
                  '#C0C0C0', '#FF69B4', '#FF1493', '#FFC0CB', '#F08080']

if len(top_countries) > len(rainbow_colors):
    step = 1 / (len(top_countries) - len(rainbow_colors) + 1)
    additional_colors = [f'hsl({i * step * 360},100%,50%)' for i in range(1, len(top_countries) - len(rainbow_colors) + 1)]
    colors = rainbow_colors + additional_colors
else:
    colors = rainbow_colors[:len(top_countries)]

# Create a horizontal bar
fig = go.Figure(go.Bar(
    x=top_countries.values,
    y=top_countries.index,
    orientation='h',
    marker=dict(color=colors)
))

# Update layout
fig.update_layout(
    title="Top 25 Countries",
    xaxis_title="Count",
    yaxis_title="Country",
    yaxis=dict(categoryorder='total ascending'),
    height=1000,
    width=1200

# Show the plot
fig.show()
```

```python
Age_Ranges = df["Age"].value_counts()

colors = ['blue', 'green', 'red', 'orange', 'purple','pink','yellow']
# Create a bar chart
fig = go.Figure([go.Bar(x=Age_Ranges.index, y=Age_Ranges.values, marker_color=colors)])

# Update layout
fig.update_layout(
    title="Age Ranges",
    xaxis_title="Age Range",
    yaxis_title="Count"
)

# Show the plot
fig.show()
```

```python
main_branch_mapping = {
    'I am a developer by profession': "Developer by profession",
    'I am not primarily a developer, but I write code sometimes as part of my work/studies': "Codes only for personal use, not a professional",
    'I used to be a developer by profession, but no longer am': "Once a Developer",
    'I code primarily as a hobby': "Code as hobby"
}
raw_df['MainBranch'] = raw_df['MainBranch'].replace(main_branch_mapping)

# Calculate the counts for each category
mb = raw_df["MainBranch"].value_counts()

colors = ['blue', 'green', 'red', 'orange', 'beige', 'violet']

# Create a horizontal bar graph
fig = go.Figure(go.Bar(
    x=mb.values,
    y=mb.index,
    orientation='h',
    marker=dict(color=colors)
))

# Update layout
fig.update_layout(
    title="Main Branch Categories",
    xaxis_title="Count",
    yaxis_title="Main Branch",
    yaxis=dict(categoryorder='total ascending'),

)

# Show the plot
fig.show()
```

```python
rw = df["RemoteWork"].value_counts()

colors = ['indigo', 'green', 'yellow', 'orange', 'pink']

# Create a pie chart
fig = go.Figure(go.Pie(
    labels=rw.index,
    values=rw.values,
    hole=0.1,
    marker=dict(colors=colors),
    textinfo='percent+label',
))

# Update layout
fig.update_layout(
    title="Remote Work",
)

# Show the plot
fig.show()
```

```python
ed_mapping = {
    'Secondary school (e.g. American high school, German Realschule or Gymnasium, etc.)': "Secondary School"
}
df['EdLevel'] = df['EdLevel'].replace(ed_mapping)

# Calculate the counts for each category
ed = df["EdLevel"].value_counts()

num_categories = len(ed)
colors = [f"hsl({h},80%,50%)" for h in [i * (360 / num_categories) for i in range(num_categories)]]

# Create a horizontal bar graph
fig = go.Figure(go.Bar(
    x=ed.values,
    y=ed.index,
    orientation='h',
    marker=dict(color=colors)
))

# Update layout
fig.update_layout(
    title="Education Level",
    xaxis_title="Count",
    yaxis_title="Education Level",
    yaxis=dict(categoryorder='total ascending'),

)

# Show the plot
fig.show()
```

```python
# Calculate the counts for each category
dt = raw_df["DevType"].value_counts()

rainbow_colors = ['#FF0000', '#FF7F00', '#FFFF00', '#00FF00', '#0000FF', '#4B0082', '#9400D3']

if len(dt) > len(rainbow_colors):
    step = 1 / (len(dt) - len(rainbow_colors) + 1)
    additional_colors = [f'hsl({i * step * 360},100%,50%)' for i in range(1, len(dt) - len(rainbow_colors) + 1)]
    colors = rainbow_colors + additional_colors
else:
    colors = rainbow_colors[:len(dt)]

# Create a horizontal bar graph
fig = go.Figure(go.Bar(
    x=dt.values,
    y=dt.index,
    orientation='h',
    marker=dict(color=colors)
))

# Update layout
fig.update_layout(
    title="Developer Types",
    yaxis=dict(categoryorder='total ascending'),
    height=1000,
    width=1200
)

# Increase the size of bars
fig.update_traces(marker=dict(line=dict(width=2)))

# Show the plot
fig.show()
```

```python
# Calculate the learning methods and their counts
CA = dict(raw_df["LearnCode"].value_counts())
total_records = raw_df["LearnCode"].shape[0]

learning_methods_percentage = {}
for methods, count in CA.items():
    for method in methods.split(";"):
        if method != "Other (please specify)":
            learning_methods_percentage[method] = learning_methods_percentage.get(method, 0) + count

# Calculate percentages
for method, count in learning_methods_percentage.items():
    learning_methods_percentage[method] = round(count * 100 / total_records, 2)

# Remove "Other (please specify)" method
learning_methods_percentage.pop("Other (please specify)", None)

# Create a pie chart
fig = go.Figure(go.Pie(
    labels=list(learning_methods_percentage.keys()),
    values=list(learning_methods_percentage.values()),
    textinfo='percent+label'
))

# Update layout
fig.update_layout(
    title="How developers learn to code"
)

# Show the plot
fig.show()
```

```python

# Calculate the platforms and their counts
CA = dict(raw_df["LearnCodeCoursesCert"].value_counts())
total_records = raw_df["LearnCode"].shape[0]

platforms_percentage = {}
for platforms, count in CA.items():
    for platform in platforms.split(";"):
        platforms_percentage[platform] = platforms_percentage.get(platform, 0) + count

# Calculate percentages
for platform, count in platforms_percentage.items():
    platforms_percentage[platform] = round(count * 100 / total_records, 2)

# Create a pie chart
fig = go.Figure(go.Pie(
    labels=list(platforms_percentage.keys()),
    values=list(platforms_percentage.values()),
    textinfo='percent+label'
))

# Update layout
fig.update_layout(
    title="Platforms for developers to get certificates"
)

# Show the plot
fig.show()
```

```python
# Calculate the languages and their counts
LCO = dict(raw_df["LanguageHaveWorkedWith"].value_counts())
total_records = raw_df["LanguageHaveWorkedWith"].shape[0]

language_mapping = {
    "Formal documentation provided by the owner of the tech": "Documentation by Tech owner",
    "Online challenges (e.g., daily or weekly coding challenges)": "Online Challenges"
}
for key, value in language_mapping.items():
    if key in LCO:
        LCO[value] = LCO.pop(key)

languages_percentage = {}
for languages, count in LCO.items():
    for language in languages.split(";"):
        languages_percentage[language] = languages_percentage.get(language, 0) + count

# Calculate percentages
for language, count in languages_percentage.items():
    languages_percentage[language] = round(count * 100 / total_records, 2)

# Sort languages by count in descending order
sorted_languages = dict(sorted(languages_percentage.items(), key=lambda item: item[1], reverse=True))

rainbow_colors = ['#FF0000', '#FF7F00', '#FFFF00', '#00FF00', '#0000FF', '#4B0082', '#9400D3', '#FF4500', '#FFD700', '#008000',
                  '#0000FF', '#000080', '#4B0082', '#800080', '#FF00FF', '#800000', '#008080', '#00FFFF', '#808000', '#808080',
                  '#C0C0C0', '#FF69B4', '#FF1493', '#FFC0CB', '#F08080']

if len(sorted_languages) > len(rainbow_colors):
    step = 1 / (len(sorted_languages) - len(rainbow_colors) + 1)
    additional_colors = [f'hsl({i * step * 360},100%,50%)' for i in range(1, len(sorted_languages) - len(rainbow_colors) + 1)]
    colors = rainbow_colors + additional_colors
else:
    colors = rainbow_colors[:len(sorted_languages)]

# Create a horizontal bar graph
fig = go.Figure(go.Bar(
    x=list(sorted_languages.values()),
    y=list(sorted_languages.keys()),
    orientation='h',
    marker=dict(color=colors)
))

# Update layout
fig.update_layout(
    title="Languages Developers Have Worked With",
    xaxis_title="Percentage",
    yaxis_title="Language",
    height=1200,
    width=1300,
    yaxis=dict(categoryorder='total ascending'),

)

# Show the plot
fig.show()
```
